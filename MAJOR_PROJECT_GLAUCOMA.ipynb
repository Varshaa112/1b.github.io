```python
import os
import pandas as pd
import shutil
from sklearn.model_selection import train_test_split

CSV_PATH = "/content/drive/MyDrive/Glaucoma_Datasets/G1020/G1020.csv"
IMAGE_DIR = "/content/drive/MyDrive/Glaucoma_Datasets/G1020/Images"
DEST_BASE = "/content/dataset"

df = pd.read_csv(CSV_PATH)

train_df, test_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df['binaryLabels'],
    random_state=42
)

def copy_images(dataframe, split):
    for i, row in enumerate(dataframe.itertuples(), start=1):
        img_name = row.imageID
        label = "glaucoma" if row.binaryLabels == 1 else "normal"

        src = os.path.join(IMAGE_DIR, img_name)
        dst = os.path.join(DEST_BASE, split, label, img_name)

        if os.path.exists(src) and not os.path.exists(dst):
            shutil.copy(src, dst)

        if i % 100 == 0:
            print(f"{split}: copied {i} images")

copy_images(train_df, "train")
copy_images(test_df, "test")

print("âœ… Image copying completed")
```

```python
import cv2
import numpy as np

IMG_SIZE = 224  # Input size for CNN models

def preprocess_image(img_path):
    """
    Reads a retinal fundus image, resizes it, and normalizes pixel values.
    This preprocessing pipeline is consistently applied during
    training, testing, and explainable AI (XAI) analysis.
    """
    # Read image
    img = cv2.imread(img_path)
    if img is None:
        raise ValueError(f"âŒ Unable to read image at path: {img_path}")

    # Convert BGR (OpenCV) to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Resize image
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

    # Normalize to [0, 1]
    img = img.astype(np.float32) / 255.0

    return img
```

```python
import os
import pandas as pd

BASE_PATH = "/content/drive/MyDrive/Glaucoma_Datasets"

# G1020 paths
csv_path = os.path.join(BASE_PATH, "G1020", "G1020.csv")
img_dir  = os.path.join(BASE_PATH, "G1020", "Images")

# Load CSV
df = pd.read_csv(csv_path)

image_paths = []
labels = []

for _, row in df.iterrows():
    img_path = os.path.join(img_dir, row["imageID"])
    if os.path.exists(img_path):
        image_paths.append(img_path)
        labels.append(row["binaryLabels"])

print("Total labeled images:", len(image_paths))
print("Glaucoma:", sum(labels))
print("Normal:", len(labels) - sum(labels))
```

```python
#Training
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    image_paths,
    labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)

print("Training images:", len(X_train))
print("Validation images:", len(X_val))
```

```python
#preprocessing
import tensorflow as tf

IMG_SIZE = 224
BATCH_SIZE = 32

def preprocess(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, (IMG_SIZE, IMG_SIZE))
    img = img / 255.0
    return img, label

train_ds = tf.data.Dataset.from_tensor_slices((X_train, y_train))
train_ds = train_ds.map(preprocess).shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

val_ds = tf.data.Dataset.from_tensor_slices((X_val, y_val))
val_ds = val_ds.map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
```

```python
#CNN
#Objective1
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Conv2D(32, 3, activation='relu', input_shape=(224,224,3)),
    layers.MaxPooling2D(),

    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),

    layers.Conv2D(128, 3, activation='relu'),
    layers.MaxPooling2D(),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
#Train CNN
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10
)
```

```python
loss, acc = model.evaluate(val_ds)
print(f"Validation Accuracy: {acc*100:.2f}%")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import VGG16, ResNet50
```

```python
#CNN
cnn_model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(2, activation='softmax')
])

cnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("âœ… CNN model ready!")
```

```python
#VGG16
vgg_base = VGG16(
    include_top=False,
    weights='imagenet',
    input_shape=(224,224,3)
)

vgg_base.trainable = False  # Freeze pretrained layers

vgg_model = models.Sequential([
    vgg_base,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(2, activation='softmax')
])

vgg_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("âœ… VGG16 model ready!")
```

```python
#ResNet50
resnet_base = ResNet50(
    include_top=False,
    weights='imagenet',
    input_shape=(224,224,3)
)

resnet_base.trainable = False

resnet_model = models.Sequential([
    resnet_base,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(2, activation='softmax')
])

resnet_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("âœ… ResNet50 model ready!")
```

```python
#Now ll combine these models to make one strong â€œensembleâ€ model.
#Step 1: Load the saved models
# ================================
# ENSEMBLE MODEL (CNN + VGG16 + ResNet50)
# ================================

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from sklearn.metrics import accuracy_score, classification_report

# -------------------------------
# 1. Load trained models
# -------------------------------
cnn_model = load_model("/content/cnn_model.h5")
vgg_model = load_model("/content/vgg16_model.h5")
resnet_model = load_model("/content/resnet50_model.h5")

print("âœ… CNN, VGG16, and ResNet50 models loaded")

# -------------------------------
# 2. Get predictions from each model
# -------------------------------
cnn_preds = cnn_model.predict(val_ds)
vgg_preds = vgg_model.predict(val_ds)
resnet_preds = resnet_model.predict(val_ds)

print("âœ… Predictions generated from all models")

# -------------------------------
# 3. Soft Voting (Average probabilities)
# -------------------------------
ensemble_preds = (cnn_preds + vgg_preds + resnet_preds) / 3

# Convert probabilities â†’ class labels
ensemble_classes = np.argmax(ensemble_preds, axis=1)

# -------------------------------
# 4. Get true labels from val_ds
# -------------------------------
true_labels = []

for _, labels in val_ds:
    true_labels.extend(labels.numpy())

true_labels = np.array(true_labels)

# -------------------------------
# 5. Evaluate Ensemble Model
# -------------------------------
ensemble_accuracy = accuracy_score(true_labels, ensemble_classes)

print("\nðŸ”¥ ENSEMBLE MODEL PERFORMANCE")
print(f"Ensemble Accuracy: {ensemble_accuracy * 100:.2f}%\n")

print("ðŸ“Š Classification Report:")
print(classification_report(
    true_labels,
    ensemble_classes,
    target_names=["Normal", "Glaucoma"]
))
```

```python
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model, Model
from tensorflow.keras.preprocessing import image

# ---------------------------
# 1. Load model
# ---------------------------
model = load_model("/content/vgg16_model.h5")
print("âœ… Model loaded")

# ---------------------------
# 2. Extract layers safely
# ---------------------------
vgg_base = model.get_layer("vgg16")                 # Functional VGG16
gap = model.layers[1]
dense1 = model.layers[2]
dropout = model.layers[3]
pred_layer = model.layers[4]

last_conv_layer = vgg_base.get_layer("block5_conv3")
print("âœ… Last conv layer:", last_conv_layer.name)

# ---------------------------
# 3. Rebuild functional graph (ðŸ”¥ THIS IS THE KEY ðŸ”¥)
# ---------------------------
inputs = vgg_base.input
x = vgg_base.output
x = gap(x)
x = dense1(x)
x = dropout(x)
outputs = pred_layer(x)

grad_model = Model(
    inputs=inputs,
    outputs=[last_conv_layer.output, outputs]
)

# ---------------------------
# 4. Load image
# ---------------------------
img_path = "/content/drive/MyDrive/Glaucoma_Datasets/G1020/Images/image_10.jpg"
img = image.load_img(img_path, target_size=(224,224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) / 255.0

# ---------------------------
# 5. Prediction
# ---------------------------
conv_out, preds = grad_model(img_array)
pred_class = np.argmax(preds[0])
print("Predicted:", "Glaucoma" if pred_class==1 else "Normal")

# ---------------------------
# 6. Grad-CAM
# ---------------------------
with tf.GradientTape() as tape:
    conv_out, preds = grad_model(img_array)
    loss = preds[:, pred_class]

grads = tape.gradient(loss, conv_out)
pooled_grads = tf.reduce_mean(grads, axis=(0,1,2))

conv_out = conv_out[0]
heatmap = tf.reduce_sum(conv_out * pooled_grads, axis=-1)
heatmap = np.maximum(heatmap, 0)
heatmap /= tf.reduce_max(heatmap)

# ---------------------------
# 7. Overlay
# ---------------------------
img_cv = cv2.imread(img_path)
img_cv = cv2.resize(img_cv, (224,224))

heatmap = cv2.resize(heatmap.numpy(), (224,224))
heatmap = np.uint8(255 * heatmap)
heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

overlay = cv2.addWeighted(img_cv, 0.6, heatmap, 0.4, 0)

# ---------------------------
# 8. Display
# ---------------------------
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Grad-CAM")
plt.imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
#Read CSV
import pandas as pd

csv_path = "/content/drive/MyDrive/Glaucoma_Datasets/G1020/G1020.csv"
df = pd.read_csv(csv_path)

print(df.head())
print(df.columns)
print(df['binaryLabels'].value_counts())
```

```python
#Load images + labels
import cv2
import numpy as np
from tqdm import tqdm

IMG_DIR = "/content/drive/MyDrive/Glaucoma_Datasets/G1020/Images_Cropped"
IMG_SIZE = 224

X = []
y_true = []

for _, row in tqdm(df.iterrows(), total=len(df)):
    img_path = f"{IMG_DIR}/{row['imageID']}"
    img = cv2.imread(img_path)

    if img is None:
        continue

    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = img / 255.0

    X.append(img)
    y_true.append(row['binaryLabels'])

X = np.array(X)
y_true = np.array(y_true)

print("âœ… Images shape:", X.shape)
print("âœ… Labels shape:", y_true.shape)
```

```python
#Load models & ensemble predict
from tensorflow.keras.models import load_model

cnn = load_model("/content/cnn_model.h5")
vgg = load_model("/content/vgg16_model.h5")
resnet = load_model("/content/resnet50_model.h5")

cnn_p = cnn.predict(X, verbose=0)
vgg_p = vgg.predict(X, verbose=0)
res_p = resnet.predict(X, verbose=0)

ensemble_probs = (cnn_p + vgg_p + res_p) / 3
ensemble_preds = np.argmax(ensemble_probs, axis=1)

print("âœ… Predictions:", ensemble_preds.shape)
```

```python
#Metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

accuracy  = accuracy_score(y_true, ensemble_preds)
precision = precision_score(y_true, ensemble_preds)
recall    = recall_score(y_true, ensemble_preds)
f1        = f1_score(y_true, ensemble_preds)

print("ðŸŽ¯ Accuracy :", accuracy)
print("ðŸŽ¯ Precision:", precision)
print("ðŸŽ¯ Recall   :", recall)
print("ðŸŽ¯ F1-score :", f1)

print("\nClassification Report:\n")
print(classification_report(
    y_true,
    ensemble_preds,
    target_names=["Normal", "Glaucoma"]
))
```
